```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io

# D liu CSV ca bn di dng mt chui


# c d liu CSV vo DataFrame ca pandas
# S dng io.StringIO  c chui nh mt tp
df = pd.read_csv("output/scored_provinces.csv")

# Ci t font h tr ting Vit (v d: 'DejaVu Sans')
# Hoc mt font khc bn  ci t c h tr ting Vit
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False  #  hin th du tr ng cch

# 1. Biu  ct: GRDP bnh qun u ngi theo Tnh/Thnh ph (Top N v Bottom N)
df_sorted_grdp_per_capita = df.sort_values('GRDP_per_capita', ascending=False)

plt.figure(figsize=(18, 10))  # Tng kch thc cho d nhn hn

# Top 15 tnh c GRDP bnh qun u ngi cao nht
plt.subplot(2, 1, 1)  # To subplot u tin
sns.barplot(x='Name', y='GRDP_per_capita',
            data=df_sorted_grdp_per_capita.head(15), palette="viridis")
plt.xlabel("Tnh/Thnh ph")
plt.ylabel("GRDP bnh qun u ngi (t VND/ngi)")
plt.title("Top 15 Tnh/Thnh ph c GRDP bnh qun u ngi cao nht")
plt.xticks(rotation=45, ha='right')  # Xoay nhn trc x  d c hn
plt.tight_layout()  # T ng iu chnh khong cch

# Bottom 15 tnh c GRDP bnh qun u ngi thp nht
plt.subplot(2, 1, 2)  # To subplot th hai
sns.barplot(x='Name', y='GRDP_per_capita', data=df_sorted_grdp_per_capita.tail(
    15).sort_values('GRDP_per_capita', ascending=True), palette="viridis_r")
plt.xlabel("Tnh/Thnh ph")
plt.ylabel("GRDP bnh qun u ngi (t VND/ngi)")
plt.title("Top 15 Tnh/Thnh ph c GRDP bnh qun u ngi thp nht")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()


# 2. Biu  ct: Tng GRDP theo Tnh/Thnh ph (Top N)
df_sorted_total_grdp = df.sort_values('Total_GRDP_BVnd', ascending=False)
plt.figure(figsize=(15, 8))
sns.barplot(x='Name', y='Total_GRDP_BVnd',
            data=df_sorted_total_grdp.head(15), palette="magma")
plt.xlabel("Tnh/Thnh ph")
plt.ylabel("Tng GRDP (t VND)")
plt.title("Top 15 Tnh/Thnh ph c Tng GRDP cao nht")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. Biu  ct: Dn s theo Tnh/Thnh ph (Top N)
df_sorted_people = df.sort_values('People', ascending=False)
plt.figure(figsize=(15, 8))
sns.barplot(x='Name', y='People',
            data=df_sorted_people.head(15), palette="coolwarm")
plt.xlabel("Tnh/Thnh ph")
plt.ylabel("Dn s (ngi)")
plt.title("Top 15 Tnh/Thnh ph c Dn s cao nht")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 4. Biu  phn tn: Tc  tng trng (%) vs GRDP bnh qun u ngi
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Rate', y='GRDP_per_capita', data=df,
                hue='score', size='People', sizes=(50, 500), palette="Set2")
plt.xlabel("Tc  tng trng (%)")
plt.ylabel("GRDP bnh qun u ngi (t VND/ngi)")
plt.title("Mi quan h gia Tc  tng trng v GRDP bnh qun u ngi")
plt.legend(title='Score / Dn s')
plt.grid(True)
plt.tight_layout()
plt.show()

# 5. Biu  phn tn: minmax_score vs GRDP_per_capita
plt.figure(figsize=(12, 7))
sns.scatterplot(x='minmax_score', y='GRDP_per_capita',
                data=df, hue='score', palette="tab10")
plt.xlabel("MinMax Score")
plt.ylabel("GRDP bnh qun u ngi (t VND/ngi)")
plt.title("Mi quan h gia MinMax Score v GRDP bnh qun u ngi")
plt.grid(True)
plt.tight_layout()
plt.show()


# 6. Biu  hp (Box plot) cho GRDP bnh qun u ngi theo 'score'
plt.figure(figsize=(10, 6))
sns.boxplot(x='score', y='GRDP_per_capita', data=df, palette="pastel")
plt.xlabel("Score")
plt.ylabel("GRDP bnh qun u ngi (t VND/ngi)")
plt.title("Phn phi GRDP bnh qun u ngi theo Score")
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()

# 7. Biu  Violin plot cho Tc  tng trng (%) theo 'score'
plt.figure(figsize=(10, 6))
sns.violinplot(x='score', y='Rate', data=df, palette="Set3")
plt.xlabel("Score")
plt.ylabel("Tc  tng trng (%)")
plt.title("Phn phi Tc  tng trng theo Score")
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()

# 8. Bn  nhit (Heatmap) th hin tng quan gia cc ct s
# Ch chn cc ct s  tnh ton tng quan
numerical_df = df[['Total_GRDP_BVnd', 'Rate',
                   'People', 'GRDP_per_capita', 'minmax_score']]
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True,
            cmap="YlGnBu", fmt=".2f", linewidths=.5)
plt.title("Bn  nhit tng quan gia cc ch s kinh t")
plt.show()
```

```python
df = pd.read_csv("output/scored_regions.csv")
df["Region"] = df["Name"]

# Plot
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(1, 2, 1)
sns.barplot(x='Region', y='Total_GRDP_BVnd', data=df, ax=ax1, palette='Blues_d')
ax1.set_title('Total GRDP (BVnd) by Region')
ax1.set_ylabel('Total GRDP (BVnd)')
ax1.set_xlabel('Region')
plt.xticks(rotation=30, ha='right')

ax2 = plt.subplot(1, 2, 2)
sns.barplot(x='Region', y='People', data=df, ax=ax2, palette='Greens_d')
ax2.set_title('Total People by Region')
ax2.set_ylabel('Total People')
ax2.set_xlabel('Region')
plt.xticks(rotation=30, ha='right')

plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='GRDP_per_capita',
            data=df, palette='Blues_d')
ax1.set_title('GRDP per People')
ax1.set_ylabel('Total GRDP (BVnd)')
ax1.set_xlabel('Region')
plt.xticks(rotation=30, ha='right')

plt.tight_layout()
plt.show()
```
