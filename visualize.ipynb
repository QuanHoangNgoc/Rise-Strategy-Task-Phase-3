```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io

# Dữ liệu CSV của bạn dưới dạng một chuỗi


# Đọc dữ liệu CSV vào DataFrame của pandas
# Sử dụng io.StringIO để đọc chuỗi như một tệp
df = pd.read_csv("output/scored_provinces.csv")

# Cài đặt font hỗ trợ tiếng Việt (ví dụ: 'DejaVu Sans')
# Hoặc một font khác bạn đã cài đặt có hỗ trợ tiếng Việt
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False  # Để hiển thị dấu trừ đúng cách

# 1. Biểu đồ cột: GRDP bình quân đầu người theo Tỉnh/Thành phố (Top N và Bottom N)
df_sorted_grdp_per_capita = df.sort_values('GRDP_per_capita', ascending=False)

plt.figure(figsize=(18, 10))  # Tăng kích thước cho dễ nhìn hơn

# Top 15 tỉnh có GRDP bình quân đầu người cao nhất
plt.subplot(2, 1, 1)  # Tạo subplot đầu tiên
sns.barplot(x='Name', y='GRDP_per_capita',
            data=df_sorted_grdp_per_capita.head(15), palette="viridis")
plt.xlabel("Tỉnh/Thành phố")
plt.ylabel("GRDP bình quân đầu người (tỷ VND/người)")
plt.title("Top 15 Tỉnh/Thành phố có GRDP bình quân đầu người cao nhất")
plt.xticks(rotation=45, ha='right')  # Xoay nhãn trục x để dễ đọc hơn
plt.tight_layout()  # Tự động điều chỉnh khoảng cách

# Bottom 15 tỉnh có GRDP bình quân đầu người thấp nhất
plt.subplot(2, 1, 2)  # Tạo subplot thứ hai
sns.barplot(x='Name', y='GRDP_per_capita', data=df_sorted_grdp_per_capita.tail(
    15).sort_values('GRDP_per_capita', ascending=True), palette="viridis_r")
plt.xlabel("Tỉnh/Thành phố")
plt.ylabel("GRDP bình quân đầu người (tỷ VND/người)")
plt.title("Top 15 Tỉnh/Thành phố có GRDP bình quân đầu người thấp nhất")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()


# 2. Biểu đồ cột: Tổng GRDP theo Tỉnh/Thành phố (Top N)
df_sorted_total_grdp = df.sort_values('Total_GRDP_BVnd', ascending=False)
plt.figure(figsize=(15, 8))
sns.barplot(x='Name', y='Total_GRDP_BVnd',
            data=df_sorted_total_grdp.head(15), palette="magma")
plt.xlabel("Tỉnh/Thành phố")
plt.ylabel("Tổng GRDP (tỷ VND)")
plt.title("Top 15 Tỉnh/Thành phố có Tổng GRDP cao nhất")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. Biểu đồ cột: Dân số theo Tỉnh/Thành phố (Top N)
df_sorted_people = df.sort_values('People', ascending=False)
plt.figure(figsize=(15, 8))
sns.barplot(x='Name', y='People',
            data=df_sorted_people.head(15), palette="coolwarm")
plt.xlabel("Tỉnh/Thành phố")
plt.ylabel("Dân số (người)")
plt.title("Top 15 Tỉnh/Thành phố có Dân số cao nhất")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 4. Biểu đồ phân tán: Tốc độ tăng trưởng (%) vs GRDP bình quân đầu người
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Rate', y='GRDP_per_capita', data=df,
                hue='score', size='People', sizes=(50, 500), palette="Set2")
plt.xlabel("Tốc độ tăng trưởng (%)")
plt.ylabel("GRDP bình quân đầu người (tỷ VND/người)")
plt.title("Mối quan hệ giữa Tốc độ tăng trưởng và GRDP bình quân đầu người")
plt.legend(title='Score / Dân số')
plt.grid(True)
plt.tight_layout()
plt.show()

# 5. Biểu đồ phân tán: minmax_score vs GRDP_per_capita
plt.figure(figsize=(12, 7))
sns.scatterplot(x='minmax_score', y='GRDP_per_capita',
                data=df, hue='score', palette="tab10")
plt.xlabel("MinMax Score")
plt.ylabel("GRDP bình quân đầu người (tỷ VND/người)")
plt.title("Mối quan hệ giữa MinMax Score và GRDP bình quân đầu người")
plt.grid(True)
plt.tight_layout()
plt.show()


# 6. Biểu đồ hộp (Box plot) cho GRDP bình quân đầu người theo 'score'
plt.figure(figsize=(10, 6))
sns.boxplot(x='score', y='GRDP_per_capita', data=df, palette="pastel")
plt.xlabel("Score")
plt.ylabel("GRDP bình quân đầu người (tỷ VND/người)")
plt.title("Phân phối GRDP bình quân đầu người theo Score")
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()

# 7. Biểu đồ Violin plot cho Tốc độ tăng trưởng (%) theo 'score'
plt.figure(figsize=(10, 6))
sns.violinplot(x='score', y='Rate', data=df, palette="Set3")
plt.xlabel("Score")
plt.ylabel("Tốc độ tăng trưởng (%)")
plt.title("Phân phối Tốc độ tăng trưởng theo Score")
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()

# 8. Bản đồ nhiệt (Heatmap) thể hiện tương quan giữa các cột số
# Chỉ chọn các cột số để tính toán tương quan
numerical_df = df[['Total_GRDP_BVnd', 'Rate',
                   'People', 'GRDP_per_capita', 'minmax_score']]
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True,
            cmap="YlGnBu", fmt=".2f", linewidths=.5)
plt.title("Bản đồ nhiệt tương quan giữa các chỉ số kinh tế")
plt.show()
```
